## Probability

- Probability is the measure of likelihood or chances that an event  will occur.
- Probability lies between '0' and '1'. If Probability of an event is '0' then it is an impossible event and If Probability of an event is '1' then it is an certain event.  

### Sample Space and Event

- Sample space is all possible outcomes.
- Event is favorable outcomes and event is always a subset of sample space.

- $ P(E) = \frac{ No.\ of\ favorable\ outcomes}{No.\ of\ total\ outcomes\ (No.\ of\ outcomes\ in\ sample\ space)} $

## Join Probability

- $ P(A\ and\ B) = P(A \cap B) $
    - Probability of Both A and B occurs.

- $ P(A\ or\ B) =  P(A \cup B) $
    - probability of either A or B or Both occurs

- If events are independent then: 
    - $ P(A\ and\ B) = P(A)P(B) $

## Conditional Probability

- conditional probability is a measure of the probability of an event occurring, given that another event is already known to have occurred.

    - $ P(B \mid A) $ = $ P(B \cap A) \over P(A)  $
    - This says probability of B occurring if A has already occurred.

## Types of Events

### Mutually Exclusive Events

- If occurrence of one event excludes occurrence of another event then the events are mutually exclusive.
- If A and B are mutually exclusive then $ P(A \cap B) =  \emptyset $

### Independent Event

- If occurrence of one event does not alter the chances of occurrence of another event then they are independent events.
- If A ans B are independent event then $ P(A \mid B) = P(A) $ and $ P(B \mid A) = P(B) $

## Bayes Theorem

- Bayes theorem gives a mathematical rule for inverting conditional probabilities, allowing the probability of a cause to be found given its effect.

    - $ {P(A\vert B)={\frac {P(B\vert A)P(A)}{P(B)}}} $ 

## Normal Distribution

- It is bell-shaped curve which shows probability of having that event is how much. this distribution is very common in nature.

```python
import numpy as np
import seaborn as sns

coin = np.random.choice([0,1], size=1000)

print(coin.mean())
sns.histplot(x=coin, kde=True)
```

```python
students = np.random.choice([0,1], size=1000, p=[.3, .7])

print(students.mean())
sns.histplot(students, kde=True)
```

```python
data = np.random.normal(loc=50, scale=10, size=1000)

print(data.mean())
print(data.std())
sns.histplot(data, kde=True)
```
