## Mean

- Mean is the average of dataset or measure of central tendency.

    - Mean ($ \mu $) = $ Sum\ of\ all\ the\ values \over No.\ of\ total\ Values $

### Means sensitivity to outliers

- The mean is highly sensitive to outliers because it incorporates every value in the dataset during its calculation, meaning extreme values can significantly distort the result.

## Median

- Median is the middle of the data set it divides dataset into two equal parts. one half contains values less than median and other half contained values greater than median.

## Mode 

- Mode is most recurring value in a dataset.

## Variance

- Variance is measure to indicate how much our data is scattered around the mean. dispertion of data.
- Variance is the squared average distance from the mean.

    - variance (v) = variance (v) = $ 1 \over n $ $ \sum_{i=1}^n (x_i - \bar{x} ) $
 
        - n = No. of values in dataset
        - $ x_i $ = $ i^{th} $ value of dataset
        - $ \bar{x} $ = mean of the values in dataset
          

- For sample dataset taken from original dataset and we want to estimate it's result on original dataset then we use :
    
    - variance (v) = $ 1 \over n-1 $ $ \sum_{i=1}^n (x_i - \bar{x} ) $

## Standard Deviation

- Standard deviation measure that indicates how much data is scattered around mean.
- Standard deviation is the average distance from the mean.

    - standard deviation ($ \sigma $) = $ \sqrt{ \frac{1}{n} \sum_{i=1}^n (x_i - \bar{x} ) } $
      
        - n = No. of values in dataset
        - $ x_i $ = $ i^{th} $ value of dataset
        - $ \bar{x} $ = mean of the values in dataset
          

### Why is Standard deviation used everywhere in machine leaning ?

- it quantifies the amount of variation or dispersion in a dataset, providing insight into how spread out the data points are from the mean

## Feature Scaling

- Feature scaling is used to reduce feature to comparable size with other features it is only applied to column
- There are two main techniques for feature scaling Normalization and Standardization.

### Normalization

- This technique reduces feature to lie between [0,1] .
- It can be calculated by
  
    - x' = $ x - x_{min} \over x_{max} - x_{min} $

### Standardization

- This technique reduces feature to lie between [-3,3] .
- It can be calculated by
    
    - x' = $ x - \mu \over \sigma $ Where,
 
        - $ \mu $ = mean of that column in dataset.
        - $ \sigma $ = standard deviation of that column in dataset. 

### Why scaling improves performance of model ?

- it reduces every feature to comparable scale and thus model gives every feature similar weightage of how they contribute in the prediction.

```python
import numpy as np
```

```python
salary = np.array([70000, 60000, 53000])
age = np.array([45, 44, 40])
```

### Mean of data using Numpy

```python
salary_mean = np.mean(salary) 
print(salary_mean)
age_mean = np.mean(age)
print(age_mean)
```

### Variance of data using Numpy

```python
salary_var = np.var(salary)
print(salary_var)
age_var = np.var(age)
print(age_var)
```

### Standard Deviation of data using Numpy

```python
salary_std = np.std(salary)
print(salary_std)
age_std = np.std(age)
print(age_std)
```

### Scaling dataset using Numpy

```python
scaled_salary = (salary - salary_mean) / (salary_std)
print(scaled_salary)
scaled_age = (age - age_mean) / (age_std)
print(scaled_age)
```
