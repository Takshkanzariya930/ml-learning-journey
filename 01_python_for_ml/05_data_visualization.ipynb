## 1. Matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np
```

### - Line Graph

```python
x = np.array([2022, 2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20, 35])

plt.plot(x, y)
```

### Add markers & Change line

```python
plt.plot(x, y, marker=".", markersize=15, markerfacecolor="yellow", linestyle="dashed")
```

```python
y1 = [12, 15, 34, 23, 11]
line_style = dict(marker=".", markersize=15, markerfacecolor="yellow", linestyle="dashed")

plt.plot(x, y, **line_style)
plt.plot(x, y1, **line_style)
```

### Add Title & Labels

```python
plt.title("Class Size", fontsize=15, family="arial", fontweight="bold", color="green")

plt.xlabel("Year", fontsize=10, fontfamily="arial")
plt.ylabel("No. Students", fontsize=10, fontfamily="arial")

plt.plot(x, y, **line_style)
plt.plot(x, y1, **line_style)
```

### Add X & Y Ticks

```python
plt.plot(x, y, **line_style)
plt.plot(x, y1, **line_style)

plt.xticks(x, [f"Y{a}" for a in x], color="orange")
plt.yticks(color="orange")

plt.show()
```

### - Grids in Line graph

```python
x = [1,2,3,4,5]
y = [13,23,11,26,23]
plt.plot(x,y)

plt.grid()
```

```python
x = [1,2,3,4,5]
y = [13,23,11,26,23]
plt.plot(x,y)

plt.grid(axis='y', linewidth=1, color="#39d3e8", linestyle="dashed")
```

### - Bar Chart

```python
fruits = ["Apple", "Orange", "Banana", "Kiwifruit", "Blueberry", "Grapes"]
values = np.array([35, 30, 10, 25, 40, 5])

plt.bar(fruits, values)
```

```python
plt.title("Weekly Fruites Consumption ")
plt.bar(fruits, values, color="green")

plt.xlabel("Fruits", fontsize=10)
plt.ylabel("Quantity", fontsize=10)
```

```python
plt.barh(fruits, values)
```

### - Pie Chart

```python
fruits = ["Apple", "Orange", "Banana", "Kiwifruit", "Blueberry", "Grapes"]
values = np.array([35, 30, 10, 25, 40, 5])

plt.pie(values, labels=fruits)
```

```python
fruits = ["Apple", "Orange", "Banana", "Kiwifruit", "Blueberry", "Grapes"]
values = np.array([35, 30, 10, 25, 40, 5])

plt.pie(values, labels=fruits, autopct="%1.1f")
```

```python
fruits = ["Apple", "Orange", "Banana", "Kiwifruit", "Blueberry", "Grapes"]
values = np.array([35, 30, 10, 25, 40, 5])
colors = ["#fb2b2b", "#fb542b", "#f7fb2b", "#9dbd5d", "#524f82","#b52a22"]

plt.pie(values, labels=fruits, autopct="%1.1f", colors=colors, explode=[0, 0.1, 0, 0, 0, 0], shadow=True, startangle=-90)

plt.show()
```

### - Scatter Plot

```python
temperature1 = [45, 62, 77, 97, 118, 122] 
humidity1 = [60, 48, 40, 30, 20, 18]

temperature2 = [50, 60, 74, 87, 145, 148] 
humidity2 = [65, 45, 30, 25, 20, 15]

plt.scatter(temperature1, humidity1)
```

```python
plt.scatter(temperature1, humidity1, color="#524f82", s=200, alpha=0.8, label="City A")
plt.scatter(temperature2, humidity2, color="#9dbd5d", s=200, alpha=0.8, label="City B")

plt.title("Weather Patterns")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.legend()
```

### - Histograms

```python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10, color="lightgreen", edgecolor="black")

plt.show()
```

### - Sub Plots

```python
x = np.array([0,1,2,3,4,5])

figure, axes= plt.subplots(2,2)

axes[0,0].plot(x, x*2, color="blue")
axes[0,0].set_title("x*2")
axes[0,1].bar(x, x**2, color="green")
axes[0,1].set_title("x**2")
axes[1,0].plot(x, x**3, color="yellow")
axes[1,0].set_title("x**3")
axes[1,1].pie(x**4, labels=x)
axes[1,1].set_title("x**4")

plt.show()
```
