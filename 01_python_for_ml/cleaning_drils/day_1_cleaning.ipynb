### - loading from new file 

```python
df = pd.read_csv("../datasets/messy_default_IMDB_dataset.csv")
```

```python
df.info()
```

### - lower cased column names and removed special character from column name

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
columns = list(df.columns)
columns[1] = "imdb_title_id"
columns[4] = 'genre'
df.columns = columns
```

```python
df.drop(["unnamed:_0", "unnamed:_8"], axis=1, inplace=True)
```

### - check duplicates

```python
df.duplicated().sum()
```

### - check null values

```python
df.isnull().sum()
```

### - drop rows containg null values because there are not fore than 4% of total data

```python
df.dropna(how="all", inplace=True)
```

```python
df.dropna(subset=["duration"], inplace=True)
```

```python
df.dropna(subset=["content_rating"], inplace=True)
```

### - set "imdb_title_id" as index to dataframe

```python
df.set_index("imdb_title_id", inplace=True)
```

### - extract years from inconsistent values in "release_year" column 

```python
df["release_year"] = df["release_year"].astype(str).str.extract(r"(\d{4})")
```

```python
df.dropna(subset=["release_year"], inplace=True)
```

### - convert "release_year" column to "int64" as year is number

```python
df["release_year"] = df["release_year"].astype("int64")
```

### - extract minutes from inconsistent values in "duration" column

```python
df["duration"] = df["duration"].astype(str).str.extract(r"(\d+)")
```

```python
df.dropna(subset=["duration"], inplace=True)
```

### - convert "duration" column to "int64" as minutes is number

```python
df["duration"] = df["duration"].astype("int64")
```

### - extract no. of votes from inconsistent values in "votes" column

```python
df["votes"] = df["votes"].astype(str).str.replace(".", "")
```

### - convert "votes" column to "int64" as votes are in number

```python
df["votes"] = df["votes"].astype("int64")
```

### - extract score from inconsistent values in "score" column

```python
df["score"] = df["score"].astype(str).str.extract(r"(\d+\.\d+)")
```

```python
df.dropna(subset=["score"], inplace=True)
```

### - convert "score" column to "float64" as score are in float

```python
df["score"] = df["score"].astype("float64")
```

### - extract income from incinsistent values in "income" column 

```python
df["income"] = df["income"].astype(str).str.extract(r"(\d+)")
```

### - convert "income" column to "int64" as incomes are in number

```python
df["income"] = df["income"].astype("int64")
```

### - convert genre in list 

```python
df["genre"] = df["genre"].str.split(',')
```

```python
df.to_csv("../datasets/clean_IMDB_dataset.csv")
```
