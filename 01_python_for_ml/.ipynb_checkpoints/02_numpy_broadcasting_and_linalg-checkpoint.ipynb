## 1. Broadcasting

- Brodcasting allows NUmPy to perform operations on arrays with different shape by virualy expanding dimentions so they matches the larger arrays's shape.
- Rules :
  - The dimentions have same size OR One of the dimentions has size of 1.

```python
import numpy as np
```

```python
array = np.array([[1,2,3]])
array2 = np.array([[1],[2],[3]])
print(array.shape)
print(array2.shape)
```

```python
print(array + array2)
print(array - array2)
print(array * array2)
print(array / array2)
print(array ** array2)
```

```python
array = np.array([[1,2,3],[4,5,6]])
array2 = np.array([[1],[2],[3]])
print(array.shape)
print(array2.shape)
```

```python
print(array * array2) # throws an error
```

```python
array = np.array([[1,2,3],[4,5,6]])
array2 = np.array([[1,2],[2,3],[3,4]])
print(array.shape)
print(array2.shape)
```

```python
print(array * array2)
```

## 2. Random Numbers

### - Random Integer

```python
rng = np.random.default_rng(seed=42)

print(rng.integers(low=1, high=7, size=(1)))
```

### - Random Float

```python
np.random.seed(seed=42)

print(np.random.uniform(low=-1, high=1, size=(3, 2, 4)))
```

### - Shuffleing an Array

```python
rng = np.random.default_rng()

array = np.array([1,2,3,4,5])
rng.shuffle(array)

print(array)

array2 = np.array([[1,2,3],[4,5,6]])
rng.shuffle(array2)

print(array2)
```

## 3. Basic Linear Algebra

```python
array = np.array([[1,2],[3,4]])
print(array)
print(array.shape)
```

### - Transpose of Matrix

```python
print(array.T)
```

### - Determinent of Matrix

```python
print(np.linalg.det(array))
```

### - Eigenvalues and EigenVectors of Matrix

```python
evals, evec = np.linalg.eig(array)
print(evals)
print(np.prod(evals))
```

```python
print(np.trace(array))
print(np.sum(np.diag(array)))
```

```python
print(np.sum(evals))
```

### - Multiplication of Matrix

```python
mat1 = np.array([[1,2],
                 [3,4]])
mat2 = np.array([[5,6],
                 [7,8]])
```

```python
print(mat1 * mat2)
```

```python
print(mat1 @ mat2)
```

### - Inverse of Matrix

```python
invmat1 = np.linalg.inv(mat1)
print(invmat1)
```

```python
print(np.round(mat1 @ invmat1))
```

### - Identity of Matrix

```python
print(np.eye(3,3))
```
