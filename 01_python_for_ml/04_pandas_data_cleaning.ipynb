## 1. Missing Values

### Droping column

```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
print(df)
```

```python
df = df.drop(columns=["Legendary", "No"])
print(df)
```

```python
print(df.isnull())
print(df.notnull())
```

```python
print(df.isnull().sum())
```

### Droping Null Values

```python
df = df.dropna(subset=["Type2"])
print(df)
print(df.info())
```

### Filling Null Values

```python
df = pd.read_csv("pokemon_data.csv")
```

```python
df = df.fillna({"Type2" : "None"})
print(df)
print(df.info())
```

```python
df['Type1'] = df['Type1'].replace({"Grass" : "GRASS", "Fire" : "FIRE"})
print(df)
```

```python
df["Name"] = df["Name"].str.lower()
print(df)
```

### Duplicates

```python
new_df = df.iloc[[0, 12, 22, 111, 123]]

df = pd.concat([df, new_df])
print(df)
```

```python
print(df.duplicated())
print(" ")
print(df.duplicated().describe())
```

```python
df = df.drop_duplicates() # Removes Duplicat Rows / Data
print(df)
```

### astype

```python
df["Legendary"] = df["Legendary"].astype(bool)
print(df)
```
