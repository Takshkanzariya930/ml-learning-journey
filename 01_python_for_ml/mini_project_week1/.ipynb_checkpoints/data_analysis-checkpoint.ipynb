```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/Messy_Employee_dataset.csv")
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_").str.strip()
```

```python
df
```

```python
df.set_index("employee_id", inplace=True)
```

```python
df
```

```python
df["phone"] = df["phone"].astype(str)
```

```python
df
```

```python
df.dropna(subset="salary", inplace=True)
```

```python
median_age = df["age"].median()
df["age"] = df["age"].fillna(median_age)
```

```python
df.info()
```

```python
df["phone"] = df["phone"].str.replace(r"/D", "", regex=True)
```

```python
df["phone"]
```

```python
# df["phone"] = df["phone"].str.slice(-10)
df["phone"] = df["phone"].str[-10:]
```

```python
df[['department', 'region']] = df["department_region"].str.split('-', expand=True) 
# expand seperates it into two column insted of one list with two elements
```

```python
 df = df.drop("department_region", axis=1)
```

```python
columns = ['first_name', 'last_name', 'age', 'department', 'region', 'status', 'join_date', 'salary','email', 'phone', 'performance_score', 'remote_work']
df = df[columns]
```

```python
df["join_date"] = pd.to_datetime(df["join_date"], errors="coerce")
```

```python
df.info()
```

```python
df.describe()
```

```python
df
```

```python
df.to_csv("datasets/cleaned_employee_dataset.csv")
```

```python
sns.stripplot(x=df["performance_score"], y=df["salary"], data=df, hue=df["age"])
```
